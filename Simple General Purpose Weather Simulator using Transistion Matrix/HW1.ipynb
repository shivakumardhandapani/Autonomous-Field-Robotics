2 (b) General Purpose Simulator for the State Transition Matrix:

The below code generates the simulator for state transition matrix

```python
import numpy as np
import scipy.linalg
import matplotlib.pyplot as plt

# Transition matrix
transition_matrix = np.array([
    [0.8, 0.2, 0.0],  # Sunny day transitions
    [0.4, 0.4, 0.2],  # Cloudy day transitions
    [0.2, 0.6, 0.2]   # Rainy day transitions
])

# Days
day = ['Sunny', 'Cloudy', 'Rainy']

# Function to generate a sequence of weather
def simulator(start_day, days):
    current_day = day.index(start_day)
    sequence = [start_day]

    for i in range(days - 1):
        next_day = np.random.choice([0, 1, 2], p = transition_matrix[current_day])
        sequence.append(day[next_day])
        current_state = next_day

    return sequence

# Example:
n=np.random.randint(100) # randomizing the number of days
first_day = np.random.choice(['Sunny', 'Cloudy', 'Rainy'])  # To choose a random starting day
weather_sequence = simulator(first_day, n+1)
print (n)
print(weather_sequence)
```

The program generates a weather simulation for the given transition matrix. The number of days are randomized and also the first day is randomized to generate an unbiased weather sequence.

(c) Stationary Distribution:

The below program executed to determine the stationary distribution of Markov Chain.


i) Initialization:
The simulation is initialized with an unbiased start day choice to avoid biased distribution by randomizing the start day.

```python
steps = 1000 # Transistions
n = 100 # Number of simulation runs performed
pi_values = np.zeros((steps, 3))

for j in range(0, n):
    start_day = 0
    current_day = start_day
    pi = np.array([0, 0, 0])
    pi[start_day] = 1
    i = 0
    while i < steps:
        current_day = np.random.choice([0, 1, 2], p = transition_matrix [current_day]) # Randomly choosing the first day
        pi[current_day] += 1
        pi_values[i] = pi / (i + 1)  # Storing the normalized pi values at each step
        i += 1

column_averages = np.mean(pi_values, axis=0)
percent_difference = np.zeros((steps, 3))

for i in range(steps):
    percent_difference[i] = 100 * np.abs(pi_values[i] - column_averages) / column_averages

print(f"Average of each column: {column_averages}")   # Calculates the overall average

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(pi_values[1:, 0], label='Sunny', color = 'orange')
plt.plot(pi_values[1:, 1], label='Cloudy', color = 'grey')
plt.plot(pi_values[1:, 2], label='Rainy', color = 'blue')
plt.xlabel('Steps')
plt.ylabel('Probability')
plt.title(f'Stationary Distribution for {steps} steps.')
plt.annotate(pi[0]/steps,
            xy=(steps-(steps/10), 0.58),
            fontsize=12)
plt.annotate(pi[1]/steps,
            xy=(steps-(steps/10), 0.2),
            fontsize=12)
plt.annotate(pi[2]/steps,
            xy=(steps-(steps/10), 0),
            fontsize=12)
plt.legend()
plt.grid(True)

print(f"Stationary Distribution: {pi/steps}")
```

(d) Closed-analytical form solution using diagonalization:

The below program executes eigen value diagonalization to find the stationary distribution of the state matrix

```python
# Eigenvalue diagonalization
eigenvalues, eigenvectors = np.linalg.eig(transition_matrix.T) # linear algebra module is used to solve distribution

# Finding the eigenvector corresponding to eigenvalue 1
index_of_one = np.argmin(np.abs(eigenvalues - 1))
stationary_vector = np.real(eigenvectors[:, index_of_one])

# Normalizing the eigenvector to sum to 1
stationary_distribution = stationary_vector / np.sum(stationary_vector)
print("Stationary Distribution (Calculated):", stationary_distribution)

# Solution derived via closed analytical form
solution = [9/14, 2/7, 1/14]
print(f"Solution provided: {solution}")
```
